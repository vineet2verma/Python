```python
import pandas as pd
import seaborn as sns
file = "d:\\RAW_DATA\\Book1.csv"

df = pd.read_csv(file)
col = ['hotelcode', 'hotelname','checkin', 'RNs','month']
col2 = ['RNs']
df1 = df[['date', 'hotelcode', 'hotelname','checkin', 'RNs','month']]
```

```python
#  FOR HIGHLIGHT MIN & MAX VALUE
df2 = df1.style.highlight_max(color='red').highlight_min(color='green',subset=col2)
df2
```

```python
# condition basic colour formatting
def fun1(val):
    if val<3: color ='red'
    elif val>=3 and val<=5: color ="yellow"
    else: color ="green"
    return f"background: {color}"
df4 = df1.style.applymap(fun1, subset=col2)
df4
```

```python
#  Style Bar
df5 = df1.style.bar(subset=['hotelcode','RNs'],align='right',color='blue')
df5
```
